```python
base_folder = "sortbyCID"
file_name="source.txt"
block_size = 100  # 每个数据块的大小

for fload_name in os.listdir("sortbyCID"):
    file_counter=0
    print(fload_name)
    path=os.path.join(base_folder,fload_name,file_name)
    #讀取索引資料
    with open(path, "r",encoding="utf-8") as file:
        lines = file.readlines()
    #分割索引資料
    sublists=[]
    for i in range(0, len(lines), block_size):
        sublists.append(lines[i:i+block_size])
    #儲存索引資料
    file_path = f"{file_counter}.txt"  # 文件路径
    file_path=os.path.join(base_folder,fload_name,file_path)
    
    for item in sublists:
        file_counter+=1
        
        file_path = f"{file_counter}.txt"  # 文件路径
        file_path=os.path.join(base_folder,fload_name,file_path)
        data=""
        for iii in item:
            data+=iii
            
        with open(file_path, "w") as file2:
            file2.write(data)
            
        file2.close()
    
    file.close()
    
   
```
